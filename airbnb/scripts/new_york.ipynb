# 2. Data Cleaning

```python
# Create a copy of df as working dataset to keep the original intact
df_clean = df.copy()
```

```python
# Convert 'price' fields (e.g., "$2,100.00") to float
df_clean['price'] = (
    df_clean['price'] \
        .astype(float) \
        .replace('[\$,]', '', regex=True)
)

# Verify conversion
print(f"Price fields data type: {df_clean['price'].dtypes}")
```

```python
# Parse dates (e.g., last_review) into datetime objects.
df_clean['last_review'] = pd.to_datetime(df_clean['last_review'], errors='coerce')

print(f"Data type for `last_review` column: {df_clean['last_review'].dtypes}")
```

```python
# Handle missing values in critical fields like reviews_per_month, host_name, and neighbourhood_group.
df_clean['reviews_per_month'] = df_clean['reviews_per_month'].fillna(0)
df_clean['host_name'] = df_clean['host_name'].fillna('Unknown')
df_clean['neighbourhood_group'] = df_clean['neighbourhood_group'].fillna('Other')
df_clean['name'] = df_clean['name'].fillna('Unnamed Listing')
df_clean.head()
```

```python
# Remove irrelevant rows (e.g., listings with zero availability or price).
initial_rows = len(df_clean)

df.dropna(subset=['price'], inplace=True)

df_clean = df_clean[
    (df_clean['price'] > 0) &
    (df_clean['availability_365'] > 0)
]

rows_removed = initial_rows - len(df_clean)

print(f'Initial rows: {initial_rows}')
print(f'Final rows after cleaning: {len(df)}')
print(f"Rows dropped (NaN price, price=0, or availability=0): {rows_removed}")
```

# 3. Data Enrichment

```python
# Create a price_per_booking column using price and minimum_nights.
df_clean['price_per_booking'] = (
    df_clean['price'] / df_clean['minimum_nights']
    ).round(2)

df_clean.head()
```

```python
#- Bucket availability into categories:
# - Full-time (availability > 300)
#- Part-time (100–300)
#- Rare (<100)

def categorize(availability_365):
    if availability_365 > 300:
        return 'Full-time'
    elif 100 <= availability_365 <= 300:
        return 'Part_time'
    else:
        return 'Rare '

df_clean['availability_bucket'] = df_clean['availability_365'].apply(categorize)
df_clean.head()
```

# 4. Data Analysis Using Pandas

#### i. What are the top 10 most expensive neighborhoods by average price?

```python
top_neighborhood = df_clean.groupby('neighbourhood') \
    .agg(avg_price = ('price', 'mean'), count=('price', 'count')) \
    .round(2) \
    .sort_values(by='avg_price', ascending=False) \
    .head(10) \
    .reset_index()

print("Top 10 most expensive neighborhoods by average_price: \n")
top_neighborhood

#print(top_neighborhood.to_markdown(index=False))
```

#### ii. What’s the average availability and price by room type?

```python
room_type_stats = df_clean.groupby('room_type', as_index=False) \
    .agg(avg_price = ('price', 'mean'), avg_availability = ('availability_365', 'mean')) \
    .round(2) 

print("Average availability and price by room type: \n")
room_type_stats

#print(room_type_stats.to_markdown(index=False))
```

#### iii. Which host has the most listings?

```python
most_listings = df_clean.groupby(['host_id', 'host_name'], as_index=False) \
    .agg(qty = ('id',"count")) \
    .sort_values(by="qty", ascending=False) \
    .head(1)

print('Host with the most listings: \n')
most_listings

#print(most_listings.to_markdown(index=False))
```

#### iv. How does average price vary across different boroughs or districts?

```python
borough_price = df_clean.groupby('neighbourhood_group') \
    .agg(avg_price = ("price", "mean"), qty = ("price", "count")) \
    .sort_values(by="avg_price", ascending=False) \
    .reset_index()

print("Average price by borough: \n")
borough_price

#print(borough_price.to_markdown(index=False))
```

#### v. How many listings have never been reviewed?

```python
never_reviewed = df_clean[df_clean['number_of_reviews'] == 0] \
    .shape[0]# count number of rows

print(f"The number of listings that have never been reviewed: {never_reviewed}")
```
